```python
import gurobipy as grb

def check_gurobi_license():
    try:
        # Create a simple model to check if Gurobi is working
        model = grb.Model("check_license")

        # Add a simple variable to the model
        x = model.addVar(name="x")

        # Set an objective function
        model.setObjective(x, grb.GRB.MAXIMIZE)

        # Optimize the model
        model.optimize()

        # If no exception is raised, the license is working
        print("Gurobi license is working!")

    except grb.GurobiError as e:
        print(f"Gurobi error: {e.message}")

    except Exception as e:
        print(f"An unexpected error occurred: {str(e)}")

# Run the check
check_gurobi_license()
```
